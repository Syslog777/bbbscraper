```python
from requests import get 
from requests import Session
import os

s = Session()
s.trust_env = False

def _request(search_text, country='USA', state='Alabama'):
    params = {
        'find_country' : country,
        'find_text' : search_text,
        'state' : state,
        'page' : 1,
        'sort' : 'Relevance',
    }
    headers = {
         'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36 Edg/90.0.818.46'
    }
    return get('https://www.bbb.org/search?', params=params, headers=headers)
# https://www.bbb.org/search?find_country=USA&find_text=PaintingCompanies&page=1&sort=Relevance
response = _request(search_text='PaintingCompanies')
response.raise_for_status()

# save responce content to file
html_filename = 'page1.html'

if os.path.exists(html_filename):
    os.remove(html_filename)
    
with open(html_filename, 'w') as fp:
    # uncomment if you want empty file
    fp.write(str(response.content))
    
# this is the end of Crawler
```

```python
from bs4 import BeautifulSoup

script_tag_contents = ''

def print_tags(soup):
    print([tag.name for tag in soup.find_all()])
def print_tag_contents(soup):
    print([str(tag) for tag in soup.find_all()])
    
with open(html_filename, 'r') as fp:
    html = fp.read()
    soup = BeautifulSoup(html, 'html.parser')
    script_tag_contents = soup.find_all(lambda tag: tag.name == 'script')
    print(script_tag_contents[1])
```

```python
def remove_all(_str, sub_str):
    return _str.replace(sub_str, "")

first_tag_removed = remove_all(str(script_tag_contents[1]), '<script>')
second_tag_removed = remove_all(first_tag_removed, '</script>')
```

```python
second_tag_removed
```

```python
businesses_list = second_tag_removed.split('\"businessName\":') # str list
```

```python
businesses_list
```

```python
businesses_nested_list = []

for string in businesses_list:
    businesses_nested_list.append(string.split(','))
```

```python
businesses_nested_list
```

```python
import csv

with open('muncher_output.csv', 'w') as csv_file:
    csv_writer = csv.writer(csv_file, delimiter=',')
    for string_list in businesses_nested_list:
        business_name = string_list[0]
        phone_number = ""
        for string in string_list:
            phone_num_index = string.find('phoneNumber')
            if phone_num_index > -1:
                for char in string:
                    if char.isdigit(): 
                        phone_number = phone_number + char
            else:
                continue
        if len(phone_number) > 0:    
            csv_writer.writerow([business_name, phone_number])    
                
    
    
```
